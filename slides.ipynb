## Memoization

Cache results of an expensive function

*also found in functools 3.4 `lrucache`*

````python
@memoize
def fib(n):
    if n in (0, 1):
        return n
    else:
        return fib(n - 1) + fib(n - 2)

fib(100)
````

## Wordcounting

````python
from toolz.curried import map
from toolz import frequencies, pipe

def stem(word):
    """ Stem word to basic form
    
    >>> stem('Hello!')
    'hello'
    """
    return word.lower().rstrip(",.!:;'-\"()_").lstrip("()'\"_")

sentence = "This cat jumped over this other cat!"

pipe(sentence, str.split, map(stem), frequencies)
````

# Introducing PyToolz

`itertoolz + functoolz + dicttoolz = toolz` :: *A collection of functions*

*  **Composable**: They interoperate due to their use of core data structures.
*  **Pure**: They don’t change their inputs or rely on external state.
*  **Lazy**: They don’t run until absolutely necessary, allowing them to support large streaming data sets.

*  **Low Tech**: They’re just functions, no syntax or magic tricks to learn
*  **Tuned**: They’re profiled and optimized
*  **Serializable**: They support common solutions for parallel computing

*  **Tested**:  100% coverage
*  **Documented**:  Comprehensive docstrings.  Also see [toolz.readthedocs.org](toolz.readthedocs.org)
*  **Mature Design**:  Inherits from long tradition of functional languages

## More Wordcounting 

Or in a more Pythonic way using generator expressions

````python
%%time
with open('tale-of-two-cities.txt') as f:
    for i in range(112):  # burn header
        next(f)
    lines = (line.strip().split() for line in f)
    words = (stem(word) for line in lines
                        for word in line)
    result = frequencies(words)
````

## More Wordcounting 

Or even in a fully verbose way

````python
def wordcount(file):
    """ Count the number of occurences of each word in a string

    >>> sentence = "This cat jumped over this other cat!"
    >>> wordcount(sentence)
    {'cat': 2, 'jumped': 1, 'other': 1, 'over': 1, 'this': 2}
    """
    word_lists = []
    for line in file:
        word_lists.append(line.strip().split())
        
    words = []
    for wl in word_lists:
        words.extend(wl)

    stemmed_words = []
    for word in words:
        stemmed_words.append(stem(word))

    counts = dict()
    for word in stemmed_words:
        if word not in counts:
            counts[word] = 1
        else:
            counts[word] += 1

    return counts

with open('tale-of-two-cities.txt') as f:
    for i in range(112):
        next(f)
    result = wordcount(f)
````
